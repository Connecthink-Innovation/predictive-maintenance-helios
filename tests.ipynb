```python
meteo_mejorada = pd.read_csv("/home/leibniz/Desktop/IHMAN/meteo_raw_data/new_meteo_mejorada.csv").drop("Unnamed: 0", axis=1)
meteo_canyelles = pd.read_csv("/home/leibniz/Desktop/IHMAN/meteo_raw_data/new_meteo_canyelles.csv").drop("Unnamed: 0", axis=1)
meteo_illora = pd.read_csv("/home/leibniz/Desktop/IHMAN/meteo_raw_data/new_meteo_illora.csv")

meteo_mejorada = first_meteo_preprocess(meteo_mejorada)
meteo_mejorada = meteo_groupby(meteo_mejorada)

meteo_canyelles = first_meteo_preprocess(meteo_canyelles)
meteo_canyelles = meteo_groupby(meteo_canyelles)

meteo_illora = first_meteo_preprocess(meteo_illora)
meteo_illora = meteo_groupby(meteo_illora)

meteo = pd.concat([meteo_mejorada, meteo_canyelles, meteo_illora])
```

```python
default_meteo = pd.DataFrame([meteo.mean(numeric_only=True)])
```

```python
current_week = default_meteo.copy()
week_1 = default_meteo.copy()
week_2 = default_meteo.copy()
week_3 = default_meteo.copy()
week_4 = default_meteo.copy()
```

```python
for serie in [current_week, week_1, week_2, week_3, week_4]:
    serie.rename(
                    columns={
                        "Dew_max_max": "Dew_max_max_",
                        "Hum_min_min": "Hum_min_min_",
                        "Wind_max_max": "Wind_max_max_",
                        "Temp_max_max": "Temp_max_max_",
                        "Temp_min_min": "Temp_min_min_",
                        "Pres_min_min": "Pres_min_min_",
                        "Wind_avg_mean": "Wind_avg_avg_",
                        "Wind_avg_std": "Wind_avg_std_",
                        "Hum_avg_mean": "Hum_avg_avg_",
                        "Dew_avg_mean": "Dew_avg_avg_",
                        "Dew_avg_std": "Dew_avg_std_",
                        "Hum_max_max": "Hum_max_max_",
                        "Temp_avg_mean": "Temp_avg_avg_",
                        "Temp_avg_std": "Temp_avg_std_",
                        "Pres_max_max": "Pres_max_max_",
                        "Pres_avg_mean": "Pres_avg_avg_",
                        "Pres_avg_std": "Pres_avg_std_",
                        "Precipitation_mean": "Precipitation_avg_",
                        "Precipitation_sum": "Precipitation_sum_",
                        "Dew_min_min": "Dew_min_min_",
                        "Wind_min_min": "Wind_min_min_",
                        "Hum_avg_std": "Hum_avg_std_"
                            },
                    inplace=True)
```

```python
current_week.columns = [col + "current_week" for col in current_week.columns]
week_1.columns = [col + "week-1" for col in week_1.columns]
week_2.columns = [col + "week-2" for col in week_2.columns]
week_3.columns = [col + "week-3" for col in week_3.columns]
week_4.columns = [col + "week-4" for col in week_4.columns]
```

```python
default_meteo = pd.concat([week_4, week_3, week_2, week_1, current_week], axis=1)
```

```python
default_meteo.to_csv("default_meteo/default_meteo.csv")
```
